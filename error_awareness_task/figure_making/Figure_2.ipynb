```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns

import pandas as pd

sns.set_style('white')
#Input data
in_file = '../data/wholesample_inhib_103.csv'
df = pd.read_csv(in_file)
```

```python
df.head()
```

```python
df.rename(columns={'mPFC': 'mPFC', 
                   'R_PP': 'R. Posterior Parietal Lobe', 
                   'L_Thal': 'L. Thalamus', 
                   'L_put': 'L. Putamen', 
                   'R_AI': 'R. Anterior Insula', 
                   'L_hipp': 'L. Hippocampus', 
                   'L_AI': 'L. Anterior Insula', 
                   'R_put_hipp': 'R. Putamen/hippocampus', 
                   'L_motor': 'L. motor area'
                   }, inplace=True)

df['accuracy'] = df['accuracy'].map({'CORRECT': 'correct', 'ERROR': 'error'})

df.head()
```

```python
pal = sns.color_palette(['indianred', 'mediumseagreen'])
```

```python
###############   L. Anterior Insula   #################

fig, ax = plt.subplots(figsize=(4, 3.5))
sns.barplot(y='L. Anterior Insula', x='accuracy', 
            data=df, ax=ax, capsize=0.1, 
            errcolor='black', ci=95, palette=pal, errwidth=2, linewidth=2.5, edgecolor="black")


def change_width(ax, new_value) :
    locs = ax.get_xticks()
    for i,patch in enumerate(ax.patches):
        current_width = patch.get_width()
        diff = current_width - new_value

        # we change the bar width
        patch.set_width(new_value)

        # we recenter the bar
        patch.set_x(patch.get_x() + diff * .5)
change_width(ax, .4)       

ax.set_ylabel('% BOLD 'r'$\Delta$', fontsize=20, fontweight='bold')
ax.set_xlabel('NOGO TRIAL', fontsize=20, fontweight='bold', labelpad=10)
ax.set_yticks([0, 0.5, 1, 1.5, 2])
ax.set_ylim(0, 2)
ax.tick_params(axis='both', labelsize=16, direction='out', length=5, left=True, width=2.5)
ax.tick_params(axis='x', labelsize=18)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_linewidth(2.5)
ax.spines['left'].set_linewidth(2.5)
#fig.set_size_inches((4.5, 4))
#for tick in ax.xaxis.get_major_ticks():
    #tick.label1.set_fontweight('bold')
fig.tight_layout()
fig.savefig(fname='../visualization/figure_2/inhib_LAI_bar', dpi=300)



###############   mPFC   #################

fig, ax = plt.subplots(figsize=(4, 3.5))
sns.barplot(y='mPFC', x='accuracy', 
            data=df, ax=ax, capsize=0.1, 
            errcolor='black', ci=95, palette=pal, errwidth=2, linewidth=2.5, edgecolor="black")
def change_width(ax, new_value) :
    locs = ax.get_xticks()
    for i,patch in enumerate(ax.patches):
        current_width = patch.get_width()
        diff = current_width - new_value

        # we change the bar width
        patch.set_width(new_value)

        # we recenter the bar
        patch.set_x(patch.get_x() + diff * .5)
change_width(ax, .4) 
ax.set_ylabel('% BOLD 'r'$\Delta$', fontsize=20, fontweight='bold')
ax.set_xlabel('NOGO TRIAL', fontsize=20, fontweight='bold', labelpad=10)
ax.set_yticks([0, 0.5, 1, 1.5, 2])
ax.set_ylim(0, 2)
ax.tick_params(axis='both', labelsize=16, direction='out', length=5, left=True, width=2.5)
ax.tick_params(axis='x', labelsize=18)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_linewidth(2.5)
ax.spines['left'].set_linewidth(2.5)
#fig.set_size_inches((4.5, 4))
#for tick in ax.xaxis.get_major_ticks():
    #tick.label1.set_fontweight('bold')
fig.tight_layout()
fig.savefig(fname='./figure_2/inhib_mPFC_bar', dpi=300)


###############   R. Putamen/hippocampus   #################

fig, ax = plt.subplots(figsize=(4, 3.5))
sns.barplot(y='R. Putamen/hippocampus', x='accuracy', 
            data=df, ax=ax, capsize=0.1, 
            errcolor='black', ci=95, palette=pal, errwidth=2, linewidth=2.5, edgecolor="black")
def change_width(ax, new_value) :
    locs = ax.get_xticks()
    for i,patch in enumerate(ax.patches):
        current_width = patch.get_width()
        diff = current_width - new_value

        # we change the bar width
        patch.set_width(new_value)

        # we recenter the bar
        patch.set_x(patch.get_x() + diff * .5)
change_width(ax, .4) 
ax.set_ylabel('% BOLD 'r'$\Delta$', fontsize=20, fontweight='bold')
ax.set_xlabel('NOGO TRIAL', fontsize=20, fontweight='bold', labelpad=10)
ax.set_yticks([-0.4, -0.2, 0.0, 0.2, 0.4])
ax.set_ylim(-0.4, 0.4)
ax.tick_params(axis='both', labelsize=16, direction='out', length=5, left=True, width=2.5)
ax.tick_params(axis='x', labelsize=18)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_linewidth(2.5)
#fig.set_size_inches((4.5, 4))
ax.axhline(0, color='black', linewidth=2.5)
#for tick in ax.xaxis.get_major_ticks():
    #tick.label1.set_fontweight('bold')
fig.tight_layout()
fig.savefig(fname='./figure_2/inhib_Rput-hipp_bar', dpi=300)

###############   R. Posterior Parietal Lobe   #################

fig, ax = plt.subplots(figsize=(4, 3.5))
sns.barplot(y='R. Posterior Parietal Lobe', x='accuracy', 
            data=df, ax=ax, capsize=0.1, 
            errcolor='black', ci=95, palette=pal, errwidth=2, linewidth=2.5, edgecolor="black")
def change_width(ax, new_value) :
    locs = ax.get_xticks()
    for i,patch in enumerate(ax.patches):
        current_width = patch.get_width()
        diff = current_width - new_value

        # we change the bar width
        patch.set_width(new_value)

        # we recenter the bar
        patch.set_x(patch.get_x() + diff * .5)
change_width(ax, .4) 
ax.set_ylabel('% BOLD 'r'$\Delta$', fontsize=20, fontweight='bold')
ax.set_xlabel('NOGO TRIAL', fontsize=20, fontweight='bold', labelpad=10)
ax.set_yticks([-0.4, -0.2, 0.0, 0.2, 0.4])
ax.set_ylim(-0.4, 0.4)
ax.tick_params(axis='both', labelsize=16, direction='out', length=5, left=True, width=2.5)
ax.tick_params(axis='x', labelsize=18)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_linewidth(2.5)
ax.axhline(0, color='black', linewidth=2.5)
#for tick in ax.xaxis.get_major_ticks():
    #tick.label1.set_fontweight('bold')
#fig.set_size_inches((5, 5))
fig.tight_layout()
fig.savefig(fname='./figure_2/inhib_RPP_bar', dpi=300)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns

import pandas as pd

sns.set_style('white')
#Input data
in_file = '../data/cog_fail_103.csv'
df = pd.read_csv(in_file)
df.head()
```

```python
df[['HIV', 'CB']]
df['HIV-status'] = df['HIV'].map({1: 'HIV-', 2: 'HIV+'})
df['CB-status'] = df['CB'].map({1: 'non-user', 2: 'user'})
df['group'] = df['HIV-status'] + ', ' + df['CB-status']

# We need a separate dataframe for each group
hiv_pos_df = df.loc[df['HIV-status'] == 'HIV+']
hiv_neg_df = df.loc[df['HIV-status'] == 'HIV-']
```

```python
# Then we make the jointplot for one of the groups only
jgrid = sns.jointplot(x='cog_fail_total', y='mPFCerr', 
                      data=hiv_pos_df, kind='reg',
                      color='purple', label='HIV positive',
                      joint_kws=dict(scatter_kws=dict(edgecolor='white', s=50)),
                      marginal_kws=dict(hist=False, kde_kws={'shade': True}),  # you can control axes with dicts
                      xlim=(-100, 500), ylim=(-100, 500), truncate=False)  # note the huge limits
fig = jgrid.fig
ax_joint = jgrid.ax_joint
ax_x = jgrid.ax_marg_x
ax_y = jgrid.ax_marg_y

# Add the HIV-negative central regression plot to the joint axis
sns.regplot(x='cog_fail_total', y='mPFCerr', data=hiv_neg_df, ax=ax_joint,
            color='green', scatter_kws=dict(edgecolor='white', s=50), label='HIV negative', truncate=False)


# Clean up the axes
ax_y.set_ylabel(None)
ax_x.set_xlabel(None)
ax_joint.set_xticks([0, 20, 40, 60, 80, 100])
ax_joint.set_xlim(-2,100)
ax_joint.set_yticks([-1, 0, 1, 2, 3, 4])
ax_joint.set_ylim(-1, 4)
ax_joint.tick_params(axis='both', which='major', labelsize=16)
# Add a legend
ax_joint.legend(fontsize=16)
ax_joint.spines['left'].set_linewidth(2.5)
ax_joint.spines['bottom'].set_linewidth(2.5)
ax_joint.set_xlabel('CFQ', fontsize=20, fontweight='bold')
ax_joint.set_ylabel(r'ERROR $\beta$', fontsize=20, fontweight='bold')
#fig.set_size_inches((8, 6))
ax_joint.tick_params(axis='both', which='major', labelsize=16, direction='out', length=5, left=True, bottom=True, width=2.5)
ax_joint.legend().set_visible(False)
fig.set_size_inches((4.5, 4))
fig.tight_layout()
fig.savefig(fname='./figure_2/mpfc_err_cogfail')
fig.show()

#######################################################################

# We need a separate dataframe for each group
hiv_pos_df = df.loc[df['HIV-status'] == 'HIV+']
hiv_neg_df = df.loc[df['HIV-status'] == 'HIV-']

# Then we make the jointplot for one of the groups only
jgrid = sns.jointplot(x='cog_fail_total', y='L_AIerr', 
                      data=hiv_pos_df, kind='reg',
                      color='purple', label='HIV positive',
                      joint_kws=dict(scatter_kws=dict(edgecolor='white', s=50)),
                      marginal_kws=dict(hist=False, kde_kws={'shade': True}),  # you can control axes with dicts
                      xlim=(-100, 500), ylim=(-100, 500), truncate=False)  # note the huge limits
fig = jgrid.fig
ax_joint = jgrid.ax_joint
ax_x = jgrid.ax_marg_x
ax_y = jgrid.ax_marg_y

# Add the HIV-negative central regression plot to the joint axis
sns.regplot(x='cog_fail_total', y='L_AIerr', data=hiv_neg_df, ax=ax_joint,
            color='green', scatter_kws=dict(edgecolor='white', s=50, linewidth=0.6), label='HIV negative', truncate=False)


# Clean up the axes
ax_y.set_ylabel(None)
ax_x.set_xlabel(None)
ax_joint.set_xticks([0, 20, 40, 60, 80, 100])
ax_joint.set_xlim(-2, 100)
ax_joint.set_yticks([-1, 0, 1, 2, 3, 4])
ax_joint.set_ylim(-1, 4)
ax_joint.spines['left'].set_linewidth(2.5)
ax_joint.spines['bottom'].set_linewidth(2.5)
ax_joint.tick_params(axis='both', which='major', labelsize=16)
# Add a legend
ax_joint.legend(fontsize=16)
ax_joint.set_xlabel('CFQ', fontsize=20, fontweight='bold')
ax_joint.set_ylabel(r'ERROR $\beta$', fontsize=20, fontweight='bold')
#fig.set_size_inches((8, 6))
ax_joint.lines[0].set_linestyle(":")
#ax_joint.lines[1].set_linestyle(":")
ax_joint.tick_params(axis='both', which='major', labelsize=16,direction='out', length=5, left=True, bottom=True, width=2.5)
ax_joint.legend().set_visible(False)
fig.set_size_inches((4.5, 4))
fig.tight_layout()
fig.savefig(fname='./figure_2/L_AI_err_cogfail')
fig.show()

##################################################

# We need a separate dataframe for each group
hiv_pos_df = df.loc[df['HIV-status'] == 'HIV+']
hiv_neg_df = df.loc[df['HIV-status'] == 'HIV-']

# Then we make the jointplot for one of the groups only
jgrid = sns.jointplot(x='cog_fail_total', y='R_put_hipperr', 
                      data=hiv_pos_df, kind='reg',
                      color='purple', label='HIV positive',
                      joint_kws=dict(scatter_kws=dict(edgecolor='white', s=50)),
                      marginal_kws=dict(hist=False, kde_kws={'shade': True}),  # you can control axes with dicts
                      xlim=(-100, 500), ylim=(-100, 500), truncate=False, ci=None)  # note the huge limits
fig = jgrid.fig
ax_joint = jgrid.ax_joint
ax_x = jgrid.ax_marg_x
ax_y = jgrid.ax_marg_y

# Add the HIV-negative central regression plot to the joint axis
sns.regplot(x='cog_fail_total', y='R_put_hipperr', data=hiv_neg_df, ax=ax_joint,
            color='green', scatter_kws=dict(edgecolor='white', s=50, linewidth=0.6), label='HIV negative', truncate=False, ci=None)


# Clean up the axes
ax_y.set_ylabel(None)
ax_x.set_xlabel(None)
ax_joint.set_xticks([0, 20, 40, 60, 80, 100])
ax_joint.set_xlim(-2,100)
ax_joint.set_yticks([-1.5, -1, -0.5, 0, 0.5, 1])
ax_joint.set_ylim(-1.5, 1)
ax_joint.tick_params(axis='both', which='major', labelsize=16)
# Add a legend
ax_joint.legend(fontsize=16)
ax_joint.spines['left'].set_linewidth(2.5)
ax_joint.spines['bottom'].set_linewidth(2.5)
ax_joint.set_xlabel('CFQ', fontsize=20, fontweight='bold')
ax_joint.set_ylabel(r'ERROR $\beta$', fontsize=20, fontweight='bold')
#fig.set_size_inches((8, 6))
ax_joint.lines[0].set_linestyle(":")
ax_joint.lines[1].set_linestyle(":")
ax_joint.tick_params(axis='both', which='major', labelsize=16,direction='out', length=5, left=True, bottom=True, width=2.5)
ax_joint.legend().set_visible(False)
fig.set_size_inches((4.5, 4))
fig.tight_layout()
fig.savefig(fname='./figure_2/R_put_err_cogfail')
fig.show()

#######################################################################

# We need a separate dataframe for each group
hiv_pos_df = df.loc[df['HIV-status'] == 'HIV+']
hiv_neg_df = df.loc[df['HIV-status'] == 'HIV-']

# Then we make the jointplot for one of the groups only
jgrid = sns.jointplot(x='cog_fail_total', y='R_PPerr', 
                      data=hiv_pos_df, kind='reg',
                      color='purple', label='HIV positive',
                      joint_kws=dict(scatter_kws=dict(edgecolor='white', s=50)),
                      marginal_kws=dict(hist=False, kde_kws={'shade': True}),  # you can control axes with dicts
                      xlim=(-100, 500), ylim=(-100, 500), truncate=False, ci=None)  # note the huge limits
fig = jgrid.fig
ax_joint = jgrid.ax_joint
ax_x = jgrid.ax_marg_x
ax_y = jgrid.ax_marg_y

# Add the HIV-negative central regression plot to the joint axis
sns.regplot(x='cog_fail_total', y='R_PPerr', data=hiv_neg_df, ax=ax_joint,
            color='green', scatter_kws=dict(edgecolor='white', s=50, linewidth=0.6), label='HIV negative', truncate=False, ci=None)


# Clean up the axes
ax_y.set_ylabel(None)
ax_x.set_xlabel(None)
ax_joint.set_xticks([0, 20, 40, 60, 80, 100])
ax_joint.set_xlim(-2,100)
ax_joint.set_yticks([-1.25, -0.75, -0.25, 0.25, 0.75])
ax_joint.set_ylim(-1.25, 0.75)
ax_joint.spines['left'].set_linewidth(2.5)
ax_joint.spines['bottom'].set_linewidth(2.5)
ax_joint.tick_params(axis='both', which='major', labelsize=16)
# Add a legend
ax_joint.legend(fontsize=16)
ax_joint.set_xlabel('CFQ', fontsize=20, fontweight='bold')
ax_joint.set_ylabel(r'ERROR $\beta$', fontsize=20, fontweight='bold')
#fig.set_size_inches((8, 6))
ax_joint.lines[0].set_linestyle(":")
ax_joint.lines[1].set_linestyle(":")
ax_joint.tick_params(axis='both', which='major', labelsize=16,direction='out', length=5, left=True, bottom=True, width=2.5)
ax_joint.legend().set_visible(False)
fig.set_size_inches((4.5, 4))
fig.tight_layout()
fig.savefig(fname='./figure_2/R_dmPC_err_cogfail')
fig.show()
```
